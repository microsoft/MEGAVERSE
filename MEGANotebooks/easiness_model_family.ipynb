```python
import pandas as pd
import scipy.stats as stats
```

```python
df = pd.read_excel("/home/t-daggarwal/MEGA_V3/MultilingualBlanketEval/MEGANotebooks/MEGA_list_of_languages 2_with_daa.xlsx")
```

```python
df
```

```python
new_df = df[df['Dataset'] == "XNLI"].fillna(0)

new_df
```

```python
new_df.columns
```

```python
rho, p_value = stats.spearmanr(new_df['GPT-3.5-Turbo Monolingual'], new_df['GPT-4 Monolingual'])
print(f"Spearman correlation coefficient: {rho:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
rho, p_value = stats.spearmanr(new_df['PaLM2 Monolingual'], new_df['GPT-4 Monolingual'])
print(f"Spearman correlation coefficient: {rho:.4f}")
print(f"P-value: {p_value:.4f}")
```
